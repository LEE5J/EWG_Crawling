```python
import requests
from bs4 import BeautifulSoup # 대소문자 주의
import time#시간지연 생성
import pymysql
search_word="*"#input()
```

```python
for num in range(1,254):
    url =f"https://www.ewg.org/skindeep/search?page={num}&per_page=36&search={search_word}&search_type=ingredients"
    req = requests.get(url)
    if req.status_code != requests.codes.ok:
        print("접속실패")
        exit()
    html = BeautifulSoup(req.text,"html.parser")
    items_name=html.select("div > main > section > div.product-tile  ")
    for item in items_name:
        title = item.select_one('p.product-name')
        data_level=item.select_one('p')
        temp=item.select_one('div.product-score')
        link = str(temp.select_one('img'))
        print(title.text,data_level.text,link[84])
    if num%55 == 0:
        time.sleep(61)
    print(num)
```
